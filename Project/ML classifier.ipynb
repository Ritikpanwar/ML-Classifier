```python
# Importing Libraries
import tkinter
import pandas as pd
import numpy as np
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
# Defining global variables
path=""
type_of_file=""
parameters=[]
models=[]
y_parameter=""
validation_size,random_state=0,0
```

```python
def set_path():
    global path
    path=entry_path.get()
    x.set(path)    
    
def set_file():
    global type_of_file
    type_of_file=y.get()
    y.set(type_of_file)
    
def set_y_param():
    global y_parameter
    y_parameter=z2.get()
    z2.set(y_parameter)

def set_validation_size():
    global validation_size
    validation_size=float(E1.get())
    vs.set(validation_size)

def set_random_state():
    global random_state
    random_state=int(E2.get())
    rs.set(random_state)

def open_dataset(path,typ):
    if typ=="csv":
        dataset=pd.read_csv(path)
        return dataset
    else:
        return None
    
def select_params():
    global parameters
    params = list()
    selec = list_para.curselection()
    for i in selec:
        j = list_para.get(i)
        params.append(j)
    parameters=params
    
def select_models():
    global models
    model = list()
    selec = list_model.curselection()
    for i in selec:
        j = list_model.get(i)
        model.append(j)
    models=model

def train_models(dataset,x_parameters,y_parameter,models,validation_size,random_state):
    """
    Train different classification Machine Learning algorithm
    PARAMETERS
    dataset: Pandas dataframe of the dataset
    x_parameters: list of labels use for training purpose
    y_parameter: to be predicted
    models: Models want to be used among LR, LDA, KNN, DT, NB, SVM
    validation_size: Validation size
    random_state: Random state
    """
    training_models=[]
    result=[]
    model_dict={"LR":LogisticRegression(),
                "LDA":LinearDiscriminantAnalysis(),
                "KNN":KNeighborsClassifier(),
                "DT":DecisionTreeClassifier(),
                "NB":GaussianNB(),
                "SVM":SVC()
               }
    X=dataset.loc[:,parameters]
    X=X.values
    Y=dataset.loc[:,y_parameter]
    Y=Y.values
    x_train,x_test,y_train,y_test=model_selection.train_test_split(X,Y,test_size=validation_size,random_state=random_state)
    for i in models:
        training_models.append((i,model_dict[i]))
    for name,mod in training_models:
        mod.fit(x_train,y_train)
        result.append([name,mod.score(x_test,y_test)])
    return result

def set_subtract(A,B):
    """
    A and B are lists
    Remove the contents of B from A and returns a new list
    """
    A,B=list(A),list(B)
    for i in B:
        if i in A:
            A.remove(i)
    return A
```

```python
# enter the path of dataset
# Let the path be : "C:/Users/ASUS/Downloads/mobile-price-classification/train.csv"

window=tkinter.Tk()
window.geometry("500x500")
window.title("Path and Type")
x=tkinter.StringVar()
label_1A=tkinter.Label(window,text="Enter the path of the file")
label_1A.grid(row=0,column=0)
entry_path=tkinter.Entry(window,textvariable=x)
entry_path.grid(row=0,column=1)
button_path=tkinter.Button(window,text="select",command=set_path)
button_path.grid(row=0,column=2)

# enter the type of file

y=y=tkinter.StringVar()
y.set("Select")
label_2A=tkinter.Label(window,text="Enter the type of the file")
label_2A.grid(row=1,column=0)
type_option=tkinter.OptionMenu(window,y,"csv","other")
type_option.grid(row=1,column=1)
button_type=tkinter.Button(window,text="submit",command=set_file)
button_type.grid(row=1,column=2)

# To go to next screen
#tkinter.Button(window,text='exit',command=terminate).grid(row=2,column=0)
tkinter.Button(window, text='next', command=window.destroy).grid(row=2,column=3)
tkinter.mainloop()

# opening dataset

dataset=open_dataset(path,type_of_file)

#Displaying head of dataset

if path!="":
    window=tkinter.Tk()
    window.title("Head of Dataset")
    head=dataset.head()
    head=np.array(head)
    rows,columns=head.shape
    labels=dataset.columns.values
    for i in range(rows): #Rows
        if i==0:
            for j in range(columns):
                b = tkinter.Label(window, text="{}".format(labels[j]))
                b.grid(row=i, column=j)
        for j in range(columns): #Columns
            b = tkinter.Label(window, text="{}".format(float(head[i][j])))
            b.grid(row=i+1, column=j)
    tkinter.Button(window,text='next',command=window.destroy).grid(row=rows+1,column=columns-1)
    tkinter.mainloop()

# Taking validation size and random state as input

window=tkinter.Tk()
window.geometry("500x500")
window.title("Validation size and Random state")
vs,rs=tkinter.StringVar(),tkinter.StringVar()
tkinter.Label(window,text="Enter Validation Size").grid(row=0,column=0)
E1=tkinter.Entry(window,textvariable=vs)
E1.grid(row=0,column=1)
tkinter.Button(window,text="Submit",command=set_validation_size).grid(row=0,column=2)
tkinter.Label(window,text="Enter Random State").grid(row=1,column=0)
E2=tkinter.Entry(window,textvariable=rs)
E2.grid(row=1,column=1)
tkinter.Button(window,text="Submit",command=set_random_state).grid(row=1,column=2)
tkinter.Button(window, text='next', command=window.destroy).grid(row=2,column=2)
tkinter.mainloop()
# Selecting parameters used for training Purpose

# x_parameters

window = tkinter.Tk()
window.geometry("500x500")
window.title("X-Parameters")
z1=tkinter.StringVar()
z1.set(" ".join(dataset.columns.values))
list_para = tkinter.Listbox(window,listvariable=z1,selectmode=tkinter.MULTIPLE)
list_para.grid(column=0, row=0, columnspan=2)
tkinter.Button(window, text="Select", command=select_params).grid(column=0, row=1)
tkinter.Button(window, text='next', command=window.destroy).grid(row=2,column=1)
tkinter.mainloop()

# y_parameter

window=tkinter.Tk()
window.geometry("500x500")
window.title("Y-Parameters")
z2=tkinter.StringVar()
z2.set("")
y_param_option=tkinter.OptionMenu(window,z2,*set_subtract(dataset.columns.values,parameters))
y_param_option.grid(row=0,column=0)
button_y_param=tkinter.Button(window,text="submit",command=set_y_param)
button_y_param.grid(row=1,column=2)
tkinter.Button(window, text='next', command=window.destroy).grid(row=2,column=2)
tkinter.mainloop()

# Selecting models

window = tkinter.Tk()
window.geometry("500x500")
window.title("Selecting Models")
m=tkinter.StringVar()
m.set("LR LDA KNN DT NB SVM")
list_model = tkinter.Listbox(window,listvariable=m,selectmode=tkinter.MULTIPLE)
list_model.grid(row=0,column=0,columnspan=2)
tkinter.Button(window, text="Select", command=select_models).grid(row=1,column=0)
tkinter.Button(window, text='next', command=window.destroy).grid(row=2,column=3)
tkinter.mainloop()

# Training models

result=train_models(dataset,parameters,y_parameter,models,validation_size,random_state)
window=tkinter.Tk()
window.title("Result")
for i in range(len(result)):
    for j in range(2):
        tkinter.Label(window,text=str(result[i][j])).grid(row=i,column=j)
tkinter.Button(window,text='close',command=window.destroy).grid(row=rows+1,column=columns-1)
tkinter.mainloop()
```
