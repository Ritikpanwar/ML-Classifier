## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),color='black',linestyle='dashdot')
plt.xlabel('x_label(x)')
plt.ylabel("y-label(sinx)")
plt.title('sinewave')
plt.show()
```

## Q.3 
- Plot a line curve with list=[2,3,4,5], use linestyle and color together for the graph and marker style as "Rotated Hexagon Marker"

```python
l=[2,3,4,5]
plt.plot(l,'b--H')
plt.xlabel('x')
plt.ylabel('y')
plt.title('line curve')
plt.show()
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
data=pd.DataFrame({'Clothing':[82,78,48,46,39],'Equipment':[48,31,51,29,18],'Accessories':[15,20,14,13,9]},index=['Delhi','Mumbai','Chennai','Kolkata','Hyderbad'])
data.index.name='Store'
data
```

```python
ind = np.arange(data.shape[0])
width = 0.3
plt.bar(ind-0.3, data.Clothing,width, label='clothing',color='black')
plt.bar(ind, data.Equipment, width,label='Equipments',color='blue')
plt.bar(ind + 0.3, data.Accessories, width,label='Equipments',color='green')
plt.grid(True)
plt.ylabel('Data')
plt.title('Sell Info')
plt.xticks(ind,('Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Hyderabad'))
plt.legend(loc='upper right')
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
data=sns.load_dataset('titanic')

data.head()
```

```python
A = sns.barplot(x='class',y='survived',hue='class',palette='Accent_r',data=data)
A.set(title = 'Passenger survived oo died', xlabel = 'Passenger') 
plt.legend(loc='upper left')                               
plt.show()
```
